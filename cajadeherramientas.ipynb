```python
import numpy as np
import math

T = 12.2
p = 753.0
Td= 11.6

T = T+273.15
Td= Td+273.15

ps=1000
Rd=287
cp=1004
E =0.622

#Temperatura potencial (theta)

def theta(T, p):
    theta=T*np.power((ps/p),Rd/cp)
    return theta


#Presión de vapor de saturación (esw)

def esw(T):
    esw=6.11*math.exp((53.49)-(6808/T)-5.09*math.log(T))
    return esw


#Presión de vapor (e): esw(Td) = e

def e(Td):
    e=6.11*math.exp((53.49)-(6808/Td)-5.09*math.log(Td))
    return e


print('Theta (Temp potencial) = ',theta(T,p),'K')
print('Presión de vapor de saturación (esw) = ', esw(T),'hPa')
print('Presión de vapor (e) = ', e(Td),'hPa') 


#Razón de mezcla (w)
def w(p, E, e):
    e=e(Td)
    w=(E*(e/(p-e)))
    return w


#Humedad relativa

def r(T, Td):
    return math.exp((T-Td)/((-0.000845)*T*Td))

#Temperatura Virual en K

def Tv(T, p, e, E):
    e=e(Td)
    return T/(1-(e/p)-((e*E/p)))


#Humedad específica (q):

def q(w):
    w=w(p, E, e)
    return w/(1.-w)



#Temperatura del nivel de condensación por ascenso en K
                   
def Tlcl(T,r):
    r=r(T, Td)
    return 1/((1/(T-55))-(math.log(r))/2840)+55


#Temperatura potencial equivalente en K

def Tep(T,p,Tlcl,w):
    Tlcl=Tlcl(T, r)
    w=w(p, E, e)
    return T*(np.power((1000/p),(0.285*(1-0.28*w))))*math.log( w*(1+(0.81*w))*((3376/Tlcl)-2.54))
    

#Altura del nivel de condensación
 
def Zc(T, Td):
    Zc=(T-Td)/8
    return Zc

                      
#Resultados

print('Humedad relativa (r) = ', r(T, Td),'%')         
print('Humedad específica (q) = ', q(w))
print('Razon de mezcla (w) = ', w(p, E, e),'' )
print('Temp virtual (Tv) = ', Tv(T, p, e, E),'K')              
print('Temp de nivel de conden (Tlcl)=',Tlcl(T, r),'K')
print('Temp potencial equivalente (Tep)= ',Tep(T,p,Tlcl,w),'K')
print('Altura de nivel de conden(Zc)= ', Zc(T, Td),'m')



```
